4. What is the purpose of the "return" statement in a Python function?
- The return statement in a Python function serves two main purposes
- 1. It Ends the Function Execution
 - When Python encounters a return statement, it immediately exits the function, skipping any code that follows it.
- 2. It Sends a Value Back to the Caller
 - The return statement is used to send data back from a function to wherever the function was called.

6. Explain the concept of generators in Python and how they are defined.
- A generator is a special type of function in Python that allows you to generate a sequence of values lazily, i.e., one value at a time, on demand, using the yield keyword.generators use yield to return a value and pause the function.

7. What are the advantages of using generators over regular functions?
- The main difference between a regular function and a generator function is that the latter uses the yield keyword to produce a series of values over time, one at a time, rather than returning a single result.

10. What is the difference between 'map()', 'reduce()' and 'filter()` functions in Python?
- 1. map(function, iterable)
 - Purpose: Transforms each element of the iterable using a function.
 - Returns: A new iterable with transformed elements.
- 2. filter(function, iterable)
 - Purpose: Filters elements based on a condition (function returns True or False).
 - Returns: Only those elements that satisfy the condition.
- 3. reduce(function, iterable)
 - reduce() is part of the functools module in Python, so it must be imported.
 - Purpose: Repeatedly applies a function to reduce the iterable to a single cumulative value.
 - Returns: A single value

```python
#1 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_even_numbers(numbers):
    even_sum = sum(num for num in numbers if num % 2 == 0)
    return even_sum
numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(numbers)
print("Sum of even numbers:", result)
```

```python
#3 Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_number(n):
    result = []
    for i in n:
        result.append(i ** 2)
    return result
square_number([1, 2, 3, 4, 5, 6, 7])
```

```python
#4 Write a Python function that checks if a given number is prime or not from 1 to 200.
def prime_no(n):
  if n <= 1:
    return False
  for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
      return False
  return True

for num in range(1, 201):
  if prime_no(num):
    print(num, " is a prime number")
  else:
    print(num, " is not a prime number")
```

```python
#5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
#fibonacci upto 1000.
def fib(n):
  a=0
  b=1
  for i in range(n):
    yield a
    a, b = b, a+b
```

```python
#8 Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [(1, 4), (3, 1), (2, 5), (4, 2)]
sorted_data = sorted(data, key = lambda x:x[1])
print(sorted_data)
```

```python
#9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius_temp = [0, 20, 37, 100]
fahrenheit_temp = list(map(lambda x: x*9/5+32, celsius_temp))
print("celsius", celsius_temp)
print("fahrenheit", fahrenheit_temp)
```

```python
#10  Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(string1):
  vowels= "aeiouAEIOU"
  a = filter(lambda x: x not in vowels, string1)
  return "".join(a)

text = "hello digital world"
result = remove_vowels(text)
print(result)
```
